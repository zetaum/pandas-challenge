```python
# Dependencies
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_student_combined_df = pd.merge(school_data_df, student_data_df, how="left", on=["school_name"])
school_student_combined_df
```

```python
#calculate total number of schools
total_number_school = school_student_combined_df['school_name'].value_counts()
#length of the count of each unique school name
num_schools = len(total_number_school)
num_schools
```

```python
#calculate total number of students
total_number_students = school_student_combined_df['Student ID'].value_counts()
#length of the count of each unique student ID
total_students = len(total_number_students)
total_students
```

```python
#calculate total budget
group_school_budget = school_student_combined_df.groupby(["school_name"]).mean()
school_budget = group_school_budget['budget'].sum()
school_budget
```

```python
#calculate average math score
avg_math_score = school_student_combined_df['math_score'].mean()
avg_math_score
```

```python
#calculate average reading score
avg_reading_score = school_student_combined_df['reading_score'].mean()
avg_reading_score
```

```python
#calculate the % of students with a passing math score (70 or greater)
passing_math_score = school_student_combined_df.loc[school_student_combined_df['math_score'] >= 70]['math_score'].count()
perc_passing_math =  (passing_math_score/total_students)*100
perc_passing_math
```

```python
#calculate the % of students with a passing reading score (70 or greater)
passing_reading_score = school_student_combined_df.loc[school_student_combined_df['reading_score'] >= 70]['reading_score'].count()
perc_passing_reading =(passing_reading_score/total_students)*100
perc_passing_reading
```

```python
#calculate the percentage of students who passed math and reading (% Overall Passing)
overall_pass = (school_student_combined_df[(school_student_combined_df['math_score'] >= 70) & (school_student_combined_df['reading_score'] >= 70)]['student_name'].count()/total_students)*100
overall_pass
```

```python
#create a dataframe to hold the above results
district_summary_df = pd.DataFrame({"Total Schools": [num_schools],
                                    "Total Students": [total_students],
                                    "Total School Budget": [school_budget],
                                    "Average Math Score": [avg_math_score],
                                    "Average Reading Score": [avg_reading_score],
                                    "% Passing Math": [perc_passing_math],
                                    "% Passing Reading":[perc_passing_reading],
                                    "% Overall Passing":[overall_pass]
})


district_summary_df
```

```python
#set new index to school name
school_type = school_data_df.set_index(["school_name"])["type"]
school_type
```

```python
#total students per school
total_students_per_school = school_student_combined_df["school_name"].value_counts()
total_students_per_school
```

```python
#total school budget
budget_per_school = school_student_combined_df.groupby(["school_name"]).mean()["budget"]
budget_per_school
```

```python
#per student budget 
per_student_budget = budget_per_school/total_students_per_school
per_student_budget
```

```python
#average math score per school
avg_math_scores = round(school_student_combined_df.groupby(["school_name"]).mean()["math_score"],2)
avg_math_scores
```

```python
#average reading score per school
avg_reading_scores = round(school_student_combined_df.groupby(["school_name"]).mean()["reading_score"],2)
avg_reading_scores
```

```python
#total students passing math after df merge
students_passing_math = school_student_combined_df[(school_student_combined_df["math_score"]>=70)]

#total students passing math per school
students_passing_math_per_school = students_passing_math.groupby(["school_name"]).count()["student_name"]

#total % students passing math per school
percentage_passing_math_per_school = round((students_passing_math_per_school/total_students_per_school)*100,2)
percentage_passing_math_per_school
```

```python
#total students passing reading after df merge
students_passing_reading = school_student_combined_df[school_student_combined_df["reading_score"]>=70]

#total students passing reading per school
students_passing_reading_per_school = students_passing_reading.groupby(["school_name"]).count()["student_name"]

#total % students passing reading per school
percentage_passing_reading_per_school = round((students_passing_reading_per_school/total_students_per_school)*100,2)
percentage_passing_reading_per_school
```

```python
#overall passing rate per school
overall_passing_rate_per_school = ((percentage_passing_math_per_school + percentage_passing_reading_per_school)/2)
#overall_passing_rate_per_school_perc = (overall_passing_rate_per_school/total_students)*100
overall_passing_rate_per_school
```

```python
#creating a summary of the school data
school_summary_df = pd.DataFrame({"School Type":school_type,
                                  "Total Students":total_students_per_school,
                                  "Total School Budget":budget_per_school,
                                  "Per Student Budget":per_student_budget,
                                  "Average Math Score":avg_math_scores,
                                  "Average Reading Score":avg_reading_scores,
                                  "% Passing Math":percentage_passing_math_per_school,
                                  "% Passing Reading":percentage_passing_reading_per_school,
                                  "% Overall Passing":percentage_passing_reading_per_school})
school_summary_df.head()
```
